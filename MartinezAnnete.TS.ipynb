## Analysis
#### Now we question, which pokemon's are most likely to be found?
    We can understand with a broad sample, that the water type is the dominant in the pokemon world, second by normal type. And the most rare type is flying. Therefore, the app is most likely to locate this type of pokemon's due to it's common apearence and dominance.
    Out of the 18 type of pokemon my analysis demonstrated that water type pokemon were most likely to be found.

## Legendary pokemons and their relation with its type
    There are fewer legendary pokemon's, which are the ones people search for the most. It may be the case that people would end up having more of these "normal" pokemons rather than Legendary ones


```python
legendaries = pokemons

pokemons_legend = pokemons["Legendary"].value_counts()
pokemons_legend.plot(kind = "pie", autopct = "%1.0f%%")
plt.title("Ratio between legenadries and non legendaries", fontsize = 18) 

Pktrue = legendaries.drop(legendaries[legendaries.Legendary == False].index)


Pklegtrue = pd.crosstab(Pktrue["Type 1"], Pktrue["Legendary"])
print(Pklegtrue)

pokemons_legend = Pktrue["Legendary"].value_counts()
print(pokemons_legend)

Pklegtrue.plot(kind = "pie", colormap = "twilight_shifted", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Types of pokemon in Legendaries", fontsize = 18)    
```

## Analysing legendaries

In the case of finding one, it is certain the legendary pokemon is a dragon or a psychic type1.

## Distribution betweeen Generations
    Having a broad sample of how many pokemon are in each gen, and how each type 1 is distributed, we can comprehend the chance of getting a pokemon of one generation and their type, and even undestand how the distribtuion changes between gens, watching how some type of pokemon are introduced more tha others

```python
PokeGen = pokemons["Generation"].value_counts()
print(PokeGen)
```

```python
PokeGenType = pd.crosstab(pokemons["Generation"], pokemons["Type 1"])
print(PokeGenType)

```

### With a table we can have a good view of the spread of pokemon in each generation where water type is still the biggest one, but in charts we can have a visual undestanding of the distribution


```python
PokeGenType.plot(kind = "bar", colormap = "jet",  stacked = True)
```

- By the charts we can see that the generation with the most pokemon is between gen 1 and 5 and gen 3 by a little bit
- We can also extrapolate that the water type is the most introduced type, in exception of **gen 6** where the **ghost type dominates**

#### We can have a better understanding of the percentages of the in pie charts the describe the relation with the firts type and in the quantity of the pokemon introduced in each gen, which the firts, fifth and third have over 160 pokemon.

```python
PkGen = pokemons

PkGen1 = PkGen.drop(PkGen[PkGen.Generation == 6].index)
PkGenO = PkGen1.drop(PkGen1[PkGen1.Generation == 5].index)
PkGenL = PkGenO.drop(PkGenO[PkGenO.Generation == 4].index)
PkGenM = PkGenL.drop(PkGenL[PkGenL.Generation == 3].index)
PkGenZ = PkGenM.drop(PkGenM[PkGenM.Generation == 2].index)

PkGen1Only = pd.crosstab(PkGenZ["Type 1"], PkGenZ["Generation"])
PkGen1Only.plot(kind = "pie", colormap = "Blues", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon Gen 1", fontsize = 18)

#Gen 2 
PkGen2 = pokemons

PkGenH = PkGen2.drop(PkGen2[PkGen2.Generation == 6].index)
PkGenB = PkGenH.drop(PkGenH[PkGenH.Generation == 5].index)
PkGenC = PkGenB.drop(PkGenB[PkGenB.Generation == 4].index)
PkGenD = PkGenC.drop(PkGenC[PkGenC.Generation == 3].index)
PkGenE = PkGenD.drop(PkGenD[PkGenD.Generation == 1].index)

PkGen2Only = pd.crosstab(PkGenE["Type 1"], PkGenE["Generation"])
PkGen2Only.plot(kind = "pie", colormap = "Greens", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon in Gen 2", fontsize = 18)

#Gen 3 
PkGen3 = pokemons

PkGenH = PkGen3.drop(PkGen3[PkGen3.Generation == 6].index)
PkGenB = PkGenH.drop(PkGenH[PkGenH.Generation == 5].index)
PkGenC = PkGenB.drop(PkGenB[PkGenB.Generation == 4].index)
PkGenD = PkGenC.drop(PkGenC[PkGenC.Generation == 2].index)
PkGenE = PkGenD.drop(PkGenD[PkGenD.Generation == 1].index)

PkGen2Only = pd.crosstab(PkGenE["Type 1"], PkGenE["Generation"])
PkGen2Only.plot(kind = "pie", colormap = "Purples", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon in Gen 3", fontsize = 18)

#Gen 4 
PkGen4 = pokemons

PkGenH = PkGen4.drop(PkGen4[PkGen4.Generation == 6].index)
PkGenB = PkGenH.drop(PkGenH[PkGenH.Generation == 5].index)
PkGenC = PkGenB.drop(PkGenB[PkGenB.Generation == 3].index)
PkGenD = PkGenC.drop(PkGenC[PkGenC.Generation == 2].index)
PkGenE = PkGenD.drop(PkGenD[PkGenD.Generation == 1].index)

PkGen2Only = pd.crosstab(PkGenE["Type 1"], PkGenE["Generation"])
PkGen2Only.plot(kind = "pie", colormap = "cool", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon in Gen 4", fontsize = 18)

#Gen 5 
PkGen5 = pokemons

PkGenH = PkGen5.drop(PkGen5[PkGen5.Generation == 6].index)
PkGenB = PkGenH.drop(PkGenH[PkGenH.Generation == 4].index)
PkGenC = PkGenB.drop(PkGenB[PkGenB.Generation == 3].index)
PkGenD = PkGenC.drop(PkGenC[PkGenC.Generation == 2].index)
PkGenE = PkGenD.drop(PkGenD[PkGenD.Generation == 1].index)

PkGen2Only = pd.crosstab(PkGenE["Type 1"], PkGenE["Generation"])
PkGen2Only.plot(kind = "pie", colormap = "twilight_shifted", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon in Gen 5", fontsize = 18)

#Gen 6 
PkGen6 = pokemons

PkGenH = PkGen6.drop(PkGen6[PkGen6.Generation == 5].index)
PkGenB = PkGenH.drop(PkGenH[PkGenH.Generation == 4].index)
PkGenC = PkGenB.drop(PkGenB[PkGenB.Generation == 3].index)
PkGenD = PkGenC.drop(PkGenC[PkGenC.Generation == 2].index)
PkGenE = PkGenD.drop(PkGenD[PkGenD.Generation == 1].index)

PkGen2Only = pd.crosstab(PkGenE["Type 1"], PkGenE["Generation"])
PkGen2Only.plot(kind = "pie", colormap = "viridis", autopct = "%1.0f%%", subplots = True, fontsize = 12)
plt.title("Type of Pokemon in Gen 6", fontsize = 18)
```

## Analysis of type 1 distribution.    
Having the percentages of each pokemon type in each generation we can determine with exactitude the relation between the quantity and type, and is easier to have a fast analysis of the dominance of type, in each gen. Having analyzed each chart, we can see a tendency to water and normal types, which we already disscused, but betweeen gen 3 and gen 6, we can observe that the tendency changes a little bit, in which **gen 3** introduced a great part in **psychic** and **steel** types; in **gen 4** we can se an increase in **electric** type, which almost wins over water type, and dominating water is **grass** type; in **gen 5** the increase is in bug and grass type. And finally in **gen 6** the types change drasticly, where the dominant type is **ghost, fairy, rock and fire.**
